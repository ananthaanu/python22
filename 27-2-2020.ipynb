# ***Matplot Library***

```python
from matplotlib import pyplot as plt
x = [2,3,5,6,7,8,9,12,13]
y = [1,2,4,5,6,9,11,19,12]
plt.plot(x,y)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,3,5,6,7,8,9,12,13]
y = [1,2,4,5,6,9,11,19,12]
plt.bar(x,y)
plt.xlabel('speed')
plt.ylabel('distance')
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,3,5,6,7,8,9,12,13]
##y = [1,2,4,5,6,9,11,19,12]
plt.pie(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
x = [2,3,5,6,7,8,9,12,13]
y = [1,2,4,5,6,9,11,19,12]
plt.scatter(x,y)
plt.show()
```
